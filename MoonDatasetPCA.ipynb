```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
X, y = make_moons(n_samples=1000, noise=0.1)
y
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
pca = PCA(n_components=1, random_state=123)
pca.fit(X)
pca_result = pca.transform(X)
pca_result
```

```python
plt.scatter(pca_result[:, 0], range(0, len(pca_result)), c=y)
```

```python
kpca = KernelPCA(n_components=1, kernel='poly', gamma=10, random_state=123)
kpca.fit(X)
kpca_result = kpca.transform(X)
kpca_result
```

```python
plt.scatter(kpca_result[:, 0], range(0, len(kpca_result)), c=y)
```
